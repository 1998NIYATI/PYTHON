```python
import pandas as pd
import numpy as np
```

```python
#dataframe
d1={"customer_id":pd.Series([1,2,3,4,5,6,8,9],"Products":pd.Series(["oven","oven","oven","tv","washing machine","oven","tv","washing machine"])}
df1 = pd.DataFrame(d1)
```

```python
d2 = {'Customer_id':pd.Series([2,4,6,7,8]),'State':pd.Series(['California','California','Texas','New York','Indiana'])}

df2 = pd.DataFrame(d2)
df2
```

```python
d2={"customer_id":pd.Series([2,4,6,7,8,9]),"State":pd.Series(["MH","Guj","Guj","MH","Guj","MH"])}
df2=pd.DataFrame(d2)
df2
```

```python
#inner join in python pandas
inner_join_df= pd.merge(df1, df2, on='customer_id', how='inner')
inner_join_df 
```

```python
#outer/full join
outer_join_df=pd.merge(df1,df2,on='customer_id',how='outer')
outer_join_df
```

```python
#left join
left_join_df=pd.merge(df1,df2,how="left")
left_join_df
```

```python
# right join in python pandas

right_join_df= pd.merge(df1, df2, on='customer_id', how='right')
right_join_df
```

```python
##OTHER TYPES OF JOINS & CONCATENATION IN PANDAS PYTHON
```

```python
# join based on index python pandas

df_index = pd.merge(df1, df2, right_index=True, left_index=True)
df_index
```

```python
# Concatenate and keep the old index python pandas

df_row = pd.concat([df1, df2])
df_row
```

```python
# Concatenate and change the index python pandas
df_row_reindex = pd.concat([df1, df2], ignore_index=True)
df_row_reindex
```
